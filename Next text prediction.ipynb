```python
''' Return the List of word of the string variable "cleaned".Here list contain repeative words'''


tokens = word_tokenize(test)
print(len(tokens))
print(tokens , end= " ") #print on same line
```

```python
import string

string.punctuation
```

```python
def stemword(word):
		x = re.findall(r'^((.*?)(लाई|ले|लागि|बाट|देखि|को|की|का|मा|माथि|कै|हरु|मै|न्ने|सँग))$', word)
		if x:
			y = re.findall(r'^((.*?)(लाई|ले|लागि|बाट|देखि|को|की|का|मा|माथि|कै|हरु|मै|न्ने|सँग))$', x[0][1])
			if y:
				return y[0][1]
			else:
				return x[0][1]
		else:
			return word
```

```python
nepali_num = re.compile(r'(०|१|२|३|४|५|६|७|८|९)+')

token= [token for token in tokens if  not nepali_num.match(token)]

exclude = string.punctuation
# exclude = set(string.punctuation) # Set of all special characters
# Remove all the special characters
tokens=[stemword(ch) for ch in token if ch not in exclude]
len(token)
```

```python
tokens
```

```python
import nltk

nltk.download('punkt')
```

```python
'''
  train_len = 3+1:
  Here 3 refer to No of Input Words we take and 1 refer to the Output words.
'''
train_len = 3+1
text_sequences = []
for i in range(train_len,len(tokens)):
    seq = tokens[i-train_len:i]
    text_sequences.append(seq)

text_sequences
```

```python
''' Convert words of text_sequences to Numerical representation '''

tokenizer = Tokenizer()

tokenizer.fit_on_texts(text_sequences)
```

```python
sequences = tokenizer.texts_to_sequences(text_sequences)

print(sequences)
```

```python
tokenizer.word_counts
```

```python
'''We will calculate the vocabulary_size by using the length extracted from tokenizer.word_index and then add 1 to it. 
    We are adding 1 because 0 is a reserved for padding and we want to start our count from 1.'''

vocabulary_size = len(tokenizer.word_counts)+1
```

```python
''' We need numpy 2D array as input and output for LSTM '''

n_sequences = np.empty([len(sequences),train_len], dtype='int32')

print(n_sequences)
```

```python
n_sequences = np.empty([len(sequences),train_len], dtype='int32')
for i in range(len(sequences)):
    n_sequences[i] = sequences[i]

print(n_sequences)
```

```python
''' Here if we do n_sequences[:-1] we get (1391, 4) where the org size was(1392,4) it mean we donot include last row.So
    To get only 3 columns n_sequences[:,:-1]'''

train_inputs = n_sequences[:,:-1]

train_inputs
```

```python
''' n_sequences[:,-1] incldue all row value with last column value '''

train_targets = n_sequences[:,-1]

print(len(train_targets))
print(train_targets)
```

```python
''' One hot Encoding of the target value '''

train_targets = to_categorical(train_targets, num_classes=vocabulary_size)
train_targets
```

# Model Building

```python
from keras.models import Sequential, load_model
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Embedding
#model = load_model("mymodel.h5")

model = Sequential()
model.add(Embedding(vocabulary_size, seq_len, input_length=seq_len))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(50,activation='relu'))
model.add(Dense(vocabulary_size, activation='softmax'))
print(model.summary())
# compile network
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(train_inputs,train_targets,epochs=500,verbose=1)
model.save("/content/drive/My Drive/Sequence Models/next_word_suggestor/LSTM/Nepmodel_stemming.h5")
```
